```
# make list from DataFrame
all_student_list = df_old.values.tolist()
print(all_student_list)
```

```



#lambda to flatten sublists
flatten = lambda l: [item for sublist in l for item in sublist]





#______EARLYSTUDENTS ____________________________________


#list slice *Early Sutdents* - grabs all items up to last three of list
early_students = all_student_list[:-3]
early_students = flatten(early_students)


#______BACKUP STUDENTS____________________________________

#list slice for *Backup Students* - excludes all items up to last three of list
backup_students = all_student_list[-3::]
backup_students = flatten(backup_students)


  


  
  

all_student_list = flatten(all_student_list)


# print('Backup', backup_students)
print('Early', early_students)
print('Backup', backup_students)
print('All', all_student_list)
```
